## 5.1.Hypothesis 1: <br>Average user ratings for Xbox One and PC platforms are the same <a class="anchor" id="fifth4-1-bullet"></a>
Tasks:
- Set the alpha threshold value yourself.
- Explain: How did you formulate the null and alternative hypotheses;
- Explain: What criteria were used to test hypotheses and why.

The supposed possibility of deviating the average user rating from the expected one in both directions, that is, the hypothesis is two-sided. Two populations need to be considered:
- Average user ratings of Xbox One platforms for the current period for making forecasts for 2017 on sales on the platform;
- Average user ratings of PC platforms for the current period for making forecasts for 2017 on sales on the platform.
<br>

The current period for compiling forecasts for 2017 on sales by platform was taken from 2006 to 2016. <br>
It uses Student's t-test, because the two compared samples are independent of each other. <br>
The alpha threshold value (critical level of statistical significance) is set to 2%. I have many omissions, I definitely have data if I set the significance level to standard at 5%. <br>

**Hypothesis:**
- **Null hypothesis H₀: Average user ratings for Xbox One and PC platforms are the same.**
- **Alternative Hypothesis H₁: Average user ratings for Xbox One and PC platforms differ.**

```python
alpha = 0.02

pc = relevant_games.query("platform == 'PC'")
xone = relevant_games.query("platform == 'XOne'")

results = st.ttest_ind(
    xone['user_score'],
    pc['user_score'],
    equal_var=False,  # Welch’s t-test, which does not imply equality of variances
    nan_policy='omit' # ignore omissions
)

print('p-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Reject the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

```python
pc['user_score'].mean()
```

```python
xone['user_score'].mean()
```

Significance level, i.e. p-value = 3%. <br>
If the average user ratings of the platforms were not the same (alternative hypothesis H₁), then the differences between the average user ratings of the platform PC = 6.7 and XOne = 6.5 could happen by chance in only 3% of cases. A 3% chance is a small chance. <br>


**H₀ null hypothesis: Average user ratings for Xbox One and PC platforms are the same - confirmed.**

## 5.2. Hypothesis 2: <br>The average user ratings for the Action and Sports genres are different <a class="anchor" id="fifth4-2-bullet"></a>
Tasks:
- Set the alpha threshold value yourself.
- Explain: How did you formulate the null and alternative hypotheses;
- Explain: What criteria were used to test hypotheses and why.

- It is necessary to consider two general aggregates: 1) games of the Action genre and 2) games of the Sports genre.
- I also took the current period for building the forecast for 2017 on sales by platform from 2006 to 2016. <br>
- I use Student's t-test, because the two compared samples are independent of each other. <br>
- The threshold alpha value (critical level of statistical significance) will set the standard value to 5%. Since only 2 passes were revealed in the data that I have by genre. <br>

**Hypotheses:**
- **H₀ null hypothesis: The average user ratings for the Action and Sports genres are the same.**
- **Alternative hypothesis H₁: Average user ratings for Action and Sports genres differ.**

```python
alpha = 0.05
action = relevant_games.query("genre == 'Action'")
sports = relevant_games.query("genre == 'Sports'")

results = st.ttest_ind(
    action['user_score'],
    sports['user_score'],
    nan_policy='omit' # игнорируем пропуски
)

print('p-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Reject the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

```python
action['user_score'].mean()
```

```python
sports['user_score'].mean()
```

The average user ratings for the Action and Sports genres are different, and a significance level of 2.64 suggests that it would not happen to get such a difference in values.

**Alternative hypothesis H₁: Average user ratings for Action and Sports genres are different - confirmed**
